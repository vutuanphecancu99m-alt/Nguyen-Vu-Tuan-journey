```markdown
### 💡 Phân biệt các loại dấu ngoặc trong Python

| Dấu | Tên gọi | Cách dùng chính | Ví dụ |
| :--- | :--- | :--- | :--- |
| **`[]`** | Square Brackets | Tạo **List** (Danh sách) hoặc truy cập chỉ số | `ds = [1, 2]`, `gd['id']` |
| **`{}`** | Curly Braces | Tạo **Dictionary** (Từ điển) hoặc dùng trong **F-string** | `{'ten': 'A'}`, `f"{gia_tri}"` |
| **`()`** | Parentheses | Gọi hàm hoặc tạo **Tuple** (Danh sách cố định) | `print("Hi")`, `toa_do = (10, 20)` |
```
